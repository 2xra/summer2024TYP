```python
#simulation of data

import numpy as np

def generate_returns(num_managers, prop_pi, alpha_mean_skilled):
    num_skilled = int(num_managers * prop_pi)
    num_unskilled = num_managers - num_skilled
    
    # Generate returns for skilled managers with alpha
    skilled_returns = np.random.normal(loc=alpha_mean_skilled, scale=1, size=num_skilled)
    
    # Generate returns for unskilled managers with no alpha
    unskilled_returns = np.random.normal(loc=0, scale=1, size=num_unskilled)
    
    # Concatenate returns of skilled and unskilled managers
    all_returns = np.concatenate((skilled_returns, unskilled_returns))
    
    return all_returns

# Parameters
num_managers = 1000
prop_pi = 0.3  # Proportion of skilled managers
alpha_mean_skilled = 0.05  # Average alpha for skilled managers

# Generate returns
returns = generate_returns(num_managers, prop_pi, alpha_mean_skilled)
```
